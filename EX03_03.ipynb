#EX03-03 空氣品質指標(AQI)
Dataset: https://data.gov.tw/dataset/40448

空氣品質相關定義：https://airtw.moenv.gov.tw/CHT/Information/Standard/AirQualityIndicatorNew.aspx

請完成以下任務：

- 留下需要的欄位（'sitename', 'county','pollutant', 'status', 'aqi'）

- 請找出除了空氣品質不良的測站（aqi > 100）

- 查詢空氣污染指標物（pollutant）有那些？

- 請找出高雄市的測站資料

- 計算各縣市的平均 AQI

```python
import pandas as pd
url = 'https://data.moenv.gov.tw/api/v2/aqx_p_432?api_key=540e2ca4-41e1-4186-8497-fdd67024ac44&limit=1000&sort=ImportDate%20desc&format=CSV'
df = pd.read_csv(url)
df.head()
```

###留下需要的欄位

```python
df = df[['sitename', 'county','pollutant', 'status', 'aqi']]
df
```

###找出空氣品質不好的測站(api>100)

```python
condition = df['aqi'] > 100
df[condition]
```

###查詢空氣汙染指標物(pollutant)有哪些?

```python
df['pollutant'].unique()
```

###請找出高雄市的測站資料

```python
condition = df['county'] == '高雄市'
df[condition]
```

###計算各縣市的平均AQI

```python
df.groupby('county')['aqi'].mean().sort_values(ascending=False)
```
