#汽車銷售數據分析報告

##摘要
本報告基於 seaborn 提供的 mpg 資料集，探討汽車燃油效率數據，分析發動機排量與燃油效率之間的關係、不同產地汽車的燃油效率比較，以及車型年份對燃油效率的影響。透過數據視覺化與統計分析，提供對汽車燃油效率的洞察，並提出可能的建議。

## 引言

###背景
在汽車工業中，燃油效率是一個重要的性能指標，影響消費者的購買決策和環境保護。分析影響燃油效率的因素可以幫助製造商優化設計，同時為消費者提供參考。

###目的
本研究的目標是透過數據分析，找出影響汽車燃油效率的主要因素，並提供可行的見解來幫助汽車製造商提升產品性能與消費者做出更明智的選擇。

###待答問題
- 發動機排量與燃油效率之間是否存在關係？
- 不同產地的汽車，燃油效率是否有顯著差異？
- 車型年份對燃油效率有何影響？

##方法

###數據來源
本研究使用 seaborn 內建的 mpg 數據集，該數據集包含一組汽車燃油效率及相關特徵數據。

###分析工具
- Python（數據處理與分析）
- Pandas（數據清理與統計分析）
- Matplotlib / Seaborn（數據視覺化）

###數據處理
- 載入 mpg 數據集並檢視數據概況。
- 探索性數據分析（EDA）以了解數據分佈與異常值。
- 繪製散點圖、長條圖、盒狀圖等，分析關鍵變數之間的關係。

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# 載入數據集
mpg = sns.load_dataset("mpg")

# 顯示數據的前五筆
print(mpg.head())

# 顯示數據基本資訊
mpg.info()
```

##結果與分析

###發動機排量與燃油效率的關係

使用散點圖來觀察發動機排量（displacement）與燃油效率（mpg）之間的關係。

```python
sns.scatterplot(x="displacement", y="mpg", data=mpg)
plt.title("Engine Displacement vs. MPG")
plt.xlabel("Engine Displacement (cubic inches)")
plt.ylabel("Miles Per Gallon (MPG)")
plt.show()
```

**分析結果**：從圖中可以看出，發動機排量越大，燃油效率通常越低，呈現負相關關係。

###不同產地汽車的燃油效率比較

我們統計各產地汽車的平均燃油效率，並使用盒狀圖呈現。

```python
sns.boxplot(x="origin", y="mpg", data=mpg)
plt.title("MPG by Car Origin")
plt.xlabel("Origin")
plt.ylabel("Miles Per Gallon (MPG)")
plt.show()
```

**分析結果**：從盒狀圖可以看出，不同產地的汽車在燃油效率上有顯著差異，某些產地的汽車整體燃油效率較高。

###車型年份對燃油效率的影響

我們使用折線圖來觀察車型年份（model_year）與平均燃油效率的變化趨勢。

```python
yearly_avg = mpg.groupby("model_year")["mpg"].mean().reset_index()
sns.lineplot(x="model_year", y="mpg", data=yearly_avg)
plt.title("Average MPG Trend by Model Year")
plt.xlabel("Model Year")
plt.ylabel("Average Miles Per Gallon (MPG)")
plt.show()
```

**分析結果**：從折線圖可以看出，隨著年份的增加，汽車的平均燃油效率有上升的趨勢，這可能與技術進步和環保法規加強有關。

###結論與建議
- 發動機排量與燃油效率：數據顯示發動機排量與燃油效率呈負相關，汽車製造商可以考慮開發更高效的小排量發動機，或在保持性能的同時優化燃油效率。

- 產地差異：不同產地的汽車在燃油效率上存在差異，消費者在選購汽車時可以考慮產地因素，而製造商可以研究高效產地的技術優勢。

- 年份趨勢：近年來的汽車燃油效率有明顯提升，這反映了技術進步和環保意識的提高。建議製造商繼續投資研發更環保的汽車技術。

- 未來研究：可以進一步探討其他因素如車重、汽缸數等對燃油效率的影響，以獲得更全面的理解。
